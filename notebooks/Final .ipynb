```python
# define a function to count matches played by a team
def count_matches(team_name, df):
    # select all rows where the 'team1' or 'team2' column contains the team name
    team_matches = df[(df['team1'] == team_name) | (df['team2'] == team_name)]

    # count the number of rows in the resulting DataFrame to get the number of matches played by the team
    num_matches = team_matches.shape[0]

    # return the result
    return num_matches

# test the function for each team
teams = ['Rajasthan Royals', 'Royal Challengers Bangalore',
       'Sunrisers Hyderabad', 'Delhi Capitals', 'Chennai Super Kings',
       'Gujarat Titans', 'Lucknow Super Giants', 'Kolkata Knight Riders',
       'Punjab Kings', 'Mumbai Indians', 'Rising Pune Supergiants',
       'Kochi Tuskers Kerala']
for team in teams:
    num_matches = count_matches(team, df)
    print('Number of matches played by', team, ':', num_matches)
```

```python
match_win_target = match_loss_target = match_win_chassing = match_loss_chassing = 0 
for i in range(len(df)) : 
    if df.tossdecision.iloc[i] == 'bat' :    # target
        if df.tosswinner.iloc[i] == df.winningteam.iloc[i] : 
            match_win_target += 1
        else :
            match_loss_target += 1
    else :          # target chase 
        if df.tosswinner.iloc[i] == df.winningteam.iloc[i] : 
            match_win_chassing += 1
        else :
            match_loss_chassing += 1
print('{} times captain choose batting option and win the match.'.format(match_win_target))
print('{} times captain choose batting option but loose the match.'.format(match_loss_target))
print('{} times captain choose fielding option and win the match.'.format(match_win_chassing))
print('{} times captain choose fielding option but loose the match.'.format(match_loss_chassing))
```
